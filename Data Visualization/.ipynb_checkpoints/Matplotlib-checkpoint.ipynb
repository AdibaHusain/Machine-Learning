```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**MATPLOTLIB BASICS**

```python
x=np.linspace(0,5,11)
x
y=x**2
y
```

```python
#LINE PLOT
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```

**SUBPLOTS**

```python
#EK PLOT AREA ME MULTIPLE PLOT BAN SAKTE HAIN
#AGAR HUMNE 2,2 DIYA MATLAB 2 ROWS AUR 2 COLUMNS
#TOH MATLAB 4 SECTIONS HONGE HUMARE PAAS TOH UNKO ALAG SE SELECT KRKE PLOT BANA SAKTE HAIN

plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
```

**OBJECT ORIENTED WAY**

```python
"""SUBPLOT ME HUMARE PAAS FIXED PLOTS THE HUM USKI POSITION CHANGE NAHI KAR SAKTE THE
BUT FIGURE ME ADD_AXES SE HUM 4 VALUES DETE HAIN START KI 2 TAB US KARTE HAIN JAB HUME
APNAGRAPH KO POSITION KARNA HOTA HAI ON TOP OF ANOTHER ONE
AUR END KI 2 VALUES TOH HUM BASE GRAPH KI AXIS BADI CHOTI KARNE KE LIYE KARTE HAIN
VALUE ALWAYS LIE BETWEEN 0-1"""
fig=plt.figure() 
axis1=fig.add_axes([0.1,0.1,.6,.8])
axis1.plot(x,y)
axis2=fig.add_axes([0.2,0.4,.3,.3])
axis2.plot(x,y)
```

**TYPES OF PLOT**

**SAVING PLOT**

```python
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("Basicplot.png") #graph ki image save ho gyi as png file
```
