```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#dataset import kara from seaborn
df=sns.load_dataset('tips')
df #isme sirf humara totalbill aur tip quantitative data hai
```

**WE USE DISTRIBUTION PLOTS TO VISUALIZE THE DISTRIBUTION OF QUANTITATIVE DATA**

```python
df['size'].unique()
```

```python
#yeh humara distribution plot/histogram plot ban gya
#Isme humare equal size ke bins create rehte hain aur us range me jo values lie krti voh aa jaati hain
#default 20 rehti hain
#ek hota hai KERNAL DENSITY ESTIMATOR (yh skewness check krta hai)
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=20,kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=20,kde=True)
#seaborn me labels wagera default bane rhte hain
```

**JOINT PLOT (FOR COMPARING 2 DATA)**

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter')
#default humara scatter hi rehta hai yeh simply values match krke points banata hai
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='kde')
#default humara scatter hi rehta hai yeh simply values match krke points banata hai
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
#default humara scatter hi rehta hai yeh simply values match krke points banata hai
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
#default humara scatter hi rehta hai yeh simply values match krke points banata hai
```

**PAIR PLOT**


```python
sns.pairplot(df)
#saari numerical values ka comparison
```

```python
sns.pairplot(df,hue='size',palette='rainbow')
#qualitative data ke basis pe colors assign kr diye for differentiation
```
