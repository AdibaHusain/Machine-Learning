```python
import seaborn as sns
import matplotlib.pyplot as plt
```

**Categorical Data Plots**
Deals with caategorical or qualitative data

```python
df=sns.load_dataset('tips')
df
```

**Count Plot**

```python
sns.countplot(x=df['sex'],hue = df['smoker']) #category ka count batata hai
#hue matlab ek aur attribute add karna for more clarity of attributes
```

**Bar Plot**

```python
 sns.barplot(x=df['sex'],y=df['total_bill'])
#x axis pe categorical data hota hai aur y pe total bills
```

You can change the estimator object to your own function, that converts a vector to a scalar Mean (default): np.mean Median: np.median Sum: np.sum Count: len (to count the number of observations) Standard Deviation: np.std Minimum: np.min Maximum: np.max

```python
import numpy as np 
sns.barplot(x = df['sex'],y = df['tip'],estimator=np.sum)
```

**BOX PLOT**

```python
sns.boxplot(x='tip',y='day',data=df,palette='rainbow')
#y me qualitative data hai aur circle jo hain voh outliers hain
```
