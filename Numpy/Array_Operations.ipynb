```python
#3--- DEEP AND SHALLOW COPY
```

```python
a=np.arange(1,21)
a
```

```python
slice=a[:5]
slice= slice *10
slice
```

```python
#EVEN AFTER MAKING THE CHANGES TO A PART OF ARRAY AS SLICE THE ORIGINAL VALUES REMAIN UNTOUCHED
#THIS IS BECAUSE ARRAY CREATES A SEPARATE SPACE AND IF YOU SLICE IT WILL BE DONE ON A SEPARATE AREA THIS IS CALLED "SHALOW COPY"
#DEEP COPY ME SLICE ME CHANGES KAROGE TO ORIGINAL ME BHI SHOW HONGE
a
```

```python
#BUT IF YOU COPY A ARRAY AND MAKE CHANGES TO THE COPIED ONE THEN THE CHANGES WILL BE SHOWN IN THE ORIGINAL ONE TOO
#THIS IS BECAUSE BOTH WILL BE STORED IN THE SAME LOCATION IN RAM
b=a
b[0]=99
b
```

```python
a
```

```python
#BUT THIS WILL BE NOT DONE IF WE USE ".copy"
b=a.copy()
b[2]=40
b
```

```python
a #no change
```

```python
#(i) STACKING ARRAY
#WE CANNOT STACK TWO MATRICES , ONLY VECTORS
```

```python
a=np.array([1,2,3,4,5])
b=np.array([6,7,8,9,10])
```

```python
np.vstack((a,b))
#VERTICALLY STACK HUA HAI MATLAB LENGTH BADHI HAI BY ADDING UP A ROW
```

```python
np.hstack((a,b))
#HORIZONTALLY STACK HUA HAI BREADTH BADHI HAI
```

```python
np.column_stack((a,b))
#column ke form me stack hue hain
```
