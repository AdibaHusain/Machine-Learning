```python
import numpy as np
import pandas as pd
```

```python
labels=['a','b','c']
lists=[10,20,30]
arr=np.array(lists)
d={1:10,2:20,3:30}
```

```python
pd.Series(lists)
```

```python
#self defined labels
#Series(values,index)
pd.Series(lists,index=labels)
```

```python
pd.Series(arr)
```

```python
#takes labels from keys and values from values
pd.Series(d)
```
