```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
```

```python
df
```

**GROUP BY**

```python
#GROUP BY CATEGORY AND SUM THE SALES
#GROUP BY IS ONLY DONE ON CATEGORICAL DATA
cat=df.groupby('Category') #always return object
for i,v in cat:
    print(i)
    print(v)
```

```python
cat=df.groupby('Category')['Sales'].sum() #selection ke liye square bracket lagate hain
cat
```

```python
#GROUP BY STORE AND CALCULATE SUM OF SALES
ad=df.groupby('Store')['Sales'].sum()
ad
```

```python
#GROUP BY MULTIPLE COLUMNS LIKE CATEGORY,STORE
c=df.groupby(['Category','Sales'])['Sales'].sum()
c
```
