```python
import numpy as np
import pandas as pd
```

```python
#1--- CREATION OF DATAFRAMES
```

```python
#Using dictionary
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
df=pd.DataFrame(data)
df
```

```python
#Using 2D list
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
col=["Name","Age","City","Salary"]
df2=pd.DataFrame(data_list,columns=col) #By only giving list the columns name would be 0,1..
df2
```

```python
#2--- SELECTION AND INDEXING OF COLUMNS
```

```python
df2['Name']
```

```python
df2[['Name','City']] #selecting 2 cols
```
