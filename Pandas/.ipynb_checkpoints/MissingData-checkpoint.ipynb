```python
#1--- FINDING MISSING DATA
```

```python
import numpy as np
import pandas as pd
```

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df=pd.DataFrame(data)
df
```

```python
df.isna()
```

```python
df.isna().sum() #sum se hum count nikal sakte hain null value ka
```

```python
df.isna().any() #atleast one null value ho
```

```python
#2--- REMOVING NULL VALUES (Works with rows)
```

```python
df.dropna() #all rows having atleast one null will be dropped
```

```python
df.dropna(thresh=3) #drop rows having atleast 3 not null values
```

```python
#3--- FILLING THE MISSING VALUES
```

```python
df.fillna(0)  
```

```python
#FILLING WITH SPECIFIC VALUE
values={'A':0,'B':100,'C':300,'D':400}  
df.fillna(value=values) #Works column wise
```

```python
df
```

```python
#FILLING WITH MEAN
```

```python
df.fillna(df.mean())
```
